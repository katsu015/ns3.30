```python
import random
```

```python
grid_num = 25 #グリッド数
ob_num = 36 # １グリッドあたりの建物数
one_six = 44 #ループの１回目と六回目だけ
pos_plus = 49 # 基本はobstacle ごとに座標は49ずつ
ob_roop = 6 #建物ループは６ごと
grid_roop = 7 # gridのループは7ごと
```

```python
random.randint(100000, 999999)
```

# 出力関数

```python
def output(posx, posy, x_plus_pos, y_plus_pos, ob_id):
    print('    <poly id="poly_'+str(ob_id) + '" type="building.yes" color="blue" fill="1" layer="-1.00" shape="'+str(posx) + '.' + str(random.randint(100000, 999999)) +','+ str(posy)+ '.' + str(random.randint(100000, 999999)) + ' ' +str(posx) + '.' + str(random.randint(100000, 999999)) +',' + str(posy + y_plus_pos)+ '.' + str(random.randint(100000, 999999)) + ' ' +str(posx + x_plus_pos) + '.' + str(random.randint(100000, 999999)) +',' + str(posy + y_plus_pos)+ '.' + str(random.randint(100000, 999999)) + ' ' +str(posx + x_plus_pos) + '.' + str(random.randint(100000, 999999)) +',' + str(posy)+ '.' + str(random.randint(100000, 999999)) + '"/>' )
```

# 1grid内描画関数(1回呼び出されると36回描画)


```python
def in_grid_draw(posx, posy, ob_id): #pos = grid内の１番左上の座標  ob_id= 左上のgrid_id
    column_count = 1 # 列
    row_count = 1 #行
    for ob_index in range (1, ob_num +1): #grid内のobstacle ごとのroop
        print(str(column_count)+','+str(row_count))
        print('start position x', posx, 'y', posy)
        if column_count == 1: #1列目
            if row_count == 1 or row_count == 6:
                output(posx,posy,one_six,one_six, ob_id)
            else:
                output(posx,posy,one_six,pos_plus, ob_id)
            posx += 45
        elif column_count == 6: #6列目
            if row_count == 1 or row_count == 6:
                output(posx,posy,one_six,one_six, ob_id)
                posy += 45
            else:
                output(posx,posy,one_six,pos_plus, ob_id)
                posy += 50
            column_count = 1
            if ob_count == ob_num: # grid内のobstacleを全描画できたとき
                ob_id+=1
                break
            else: #列の末端
                posx -= 245
                ob_id += 1
                row_count +=1
                continue

        else: #2,3,4,5列目
            if row_count == 1 or row_count == 6:
                output(posx,posy,pos_plus,one_six, ob_id)
            else:
                output(posx,posy,pos_plus,pos_plus, ob_id)
            posx += 50
        column_count += 1
        ob_id += 1
```

```python
posx = 10
posy = 10
grid_count = 1
ob_count = 1
ob_id = 0 #建物のID
column_count = 1 # 列
row_count = 1 #行

for grid_index in range (1, grid_num + 1): #grid ごとのroop
    print('-----------------------------------------------------------position posx', posx, 'posy', posy)
    if column_count == 7:
        in_grid_draw(posx,posy,ob_id) # 描画開始
        posx -= 1800
        posy += 300
        column_count = 1
    else:
        print('posx', posx, 'posy', posy)
        in_grid_draw(posx,posy,ob_id) # 描画開始
        posx += 300
        column_count += 1
    ob_id = ob_id + 36
    grid_count += 1
```
