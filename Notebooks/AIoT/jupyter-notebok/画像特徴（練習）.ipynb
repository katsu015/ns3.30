```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import seaborn
from PIL import Image
```

```python
img1 = cv2.imread('color.jpg')
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
akaze = cv2.AKAZE_create()
kp1 = akaze.detect(gray1)
img1_akaze = cv2.drawKeypoints(gray1, kp1, None, flags=4)
```

```python
plt.imshow(img1_akaze)
```

```python
img  = cv2.imread('color.jpg')
print(type(img))
# <class 'numpy.ndarray'>

print(img.shape)
# (225, 400, 3)

img_rotate_90_clockwise = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

img_rotate_90_counterclockwise = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

img_rotate_180 = cv2.rotate(img, cv2.ROTATE_180)

print(img_rotate_90_clockwise.shape, img_rotate_90_counterclockwise.shape, img_rotate_180.shape)
```
