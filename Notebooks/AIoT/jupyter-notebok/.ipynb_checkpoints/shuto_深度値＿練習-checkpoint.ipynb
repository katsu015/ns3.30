```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('data/image_folder/apple/apple_color_0.jpg')
img.shape #縦～ピクセル　横～ピクセル のRGB画像
```

```python
for col in img: 
    print(col.shape)
    
```

```python
def get_depth_feature(img):
    depth_dict = {}
    depth_list = []
    MAX_DEPTH = 400
    for col in img:  #縦要素１行ずつ取得
        for num in col:
            if num not in depth_dict:
                depth_dict[num] = 1
            else:
                depth_dict[num] += 1
    
    for num in range(5, MAX_DEPTH):
        if num not in depth_dict:
            depth_list.append(0)
        else:
            depth_list.append(depth_dict[num])
    
    return depth_list
```

```python
depth = get_depth_feature(img)
```

```python
plt.figure(figsize=(12, 5))
plt.plot(depth, color='red', label='depth')
plt.legend()
plt.show()
```
