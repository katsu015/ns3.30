# 輪郭抽出 バウンディングボックス


findContoursの返り値contoursは輪郭毎の座標組、hierarchyはその輪郭の階層構造が入る
https://axa.biopapyrus.jp/ia/opencv/detect-contours.html

```python
img = cv2.imread('data/image_folder/apple/apple_color_0.jpg')
img_copy = img.copy()
img_HSV = cv2.cvtColor(img_copy, cv2.COLOR_BGR2HSV)
img_HSV = cv2.GaussianBlur(img_HSV, (9, 9), 3)

#アップルパンを検出
# detect tulips
img_H, img_S, img_V = cv2.split(img_HSV)
_thre, img_apple = cv2.threshold(img_H, 60, 255, cv2.THRESH_BINARY)
cv2.imwrite('data/image_folder/apple/apple_mask.jpg', img_apple)


contours, hierarchy = cv2.findContours(img_apple, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
xList = []
yList = []
wList = []
hList = []


for i in range(0, len(contours)):
    if len(contours[i]) > 0:

        # remove small objects
        if cv2.contourArea(contours[i]) < 500:
            continue

        rect = contours[i]
        x, y, w, h = cv2.boundingRect(rect)
        cv2.rectangle(img_copy, (x, y), (x + w, y + h), (0, 0, 0), 10)
        cv2.circle(img_copy, (x,y), 9, (255,255,255), thickness= -1)
        cv2.circle(img_copy, (x+w,y), 9, (255,255,255), thickness= -1)
        cv2.circle(img_copy, (x,y+h), 9, (255,255,255), thickness= -1)
        cv2.circle(img_copy, (x+w,y+h), 9, (255,255,255), thickness= -1)
        xList.append(x)
        yList.append(y)
        wList.append(w)
        hList.append(h)

# save
cv2.imwrite('data/image_folder/apple/apple_boundingbox.jpg', img_copy)
RGB_img = plt.imshow(cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB))
plt.show(RGB_img)
```

# 色情報
collections.Counter    https://note.nkmk.me/python-collections-counter/

```python
def count_color(image):
    color_list = []
    for i in range(3):
        color = []
        color_dict = collections.Counter(image[:, :, i].ravel()) #色ごとの画素値を１次元リストに
        for num in range(256):
            if not num in color_dict:#numという画素値がない場合0にする
                color_dict[num] = 0
        for index in sorted(color_dict.keys()): #0からソート
            color.append(color_dict[index]) #0から累計の値を格納
        color_list.append(color) #これを３チャンネル分
    return color_list

def count_hsv(image):
    hsv_list = []
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    h,s,v = cv2.split(hsv)
    for i, data in enumerate([h, s, v]):
        sort_data = []
        end = 256
        data_dict = collections.Counter(data.ravel())
        if i == 0:
            end = 180
        for num in range(end):
            if not num in data_dict:
                data_dict[num] = 0
        for index in sorted(data_dict.keys()):
            sort_data.append(data_dict[index])
        hsv_list.append(sort_data)
    return hsv_list

def view_color(result_list):
    plt.figure(figsize=(12, 5))
    plt.plot(result_list[0], color='red', label='Red')
    plt.plot(result_list[1], color='green', label='Green')
    plt.plot(result_list[2], color='blue', label='Blue')
    plt.legend()
    plt.show()
    
def get_color_feature(paths):
    result_dict = {}
    for path in paths:
        img = cv2.imread(path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        name = os.path.split(path)
        name = os.path.splitext(name[1])
        name = name[0].split('_train')[0]
        

        #RGB ３列の配列形式を１列に変換
        RGB = list(itertools.chain.from_iterable(count_color(img)))
        HSV = list(itertools.chain.from_iterable(count_hsv(img)))
        #RGBとHSVを一列につなげる
        color_feature = RGB + HSV
        result_dict[name] = color_feature
    df = pd.DataFrame.from_dict(result_dict, orient='index')
    return df
    
    
```

# 学習
https://gotutiyan.hatenablog.com/entry/2020/09/08/122621

```python
train_y = train_data.index #indexを格納　目的変数 一次元リスト
le = LabelEncoder() #LavelEncoderの宣言
le = le.fit(train_y) #ラベルとラベルIDの対応付け
train_y = le.transform(train_y)  # 対応付けに従って変換する
goods_list = list(le.classes_)
goods_dict = {} #IDとラベルを対応付けした辞書配列を作成
for i, name in enumerate(goods_list):
    goods_dict[i] = name
    
y_train = train_y
X_train = df.values
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

# テスト

```python
test_path = glob.glob('data/pan_test_data/test/*')
test = get_color_feature(test_path)
predicts = knn.predict(test)
match = 0
all_len = len(predicts)
    
for i, pre in enumerate(predicts):
    acc_name = os.path.split(test_path[i])
    acc_name = os.path.splitext(acc_name[1])
    acc_name = acc_name[0].split('_test')[0] #実際のラベル
    pre_name = goods_dict[pre]
    text_acc = 'acc: {}'.format(acc_name)
    text_name = 'pre: {}'.format(pre_name)
    text_bar = '-'*30+'\n'
    print('acc:', acc_name)
    print('pre:', pre_name)
    if acc_name == pre_name:
        match += 1
    print('-'*30)
    with open('color_result.txt', 'a') as f:
        f.write(text_acc+'\n')
        f.write(text_name+'\n')
        f.write(text_bar)
    
    
print('Accuracy Rate:', str(round(match/all_len*100, 1))+'%', '('+str(match)+'/'+str(all_len)+')')
text_result = 'Accuracy Rate: {} {}'.format(str(round(match/all_len*100, 1))+'%', '('+str(match)+'/'+str(all_len)+')')
with open('HSV_result.txt', 'a') as f:
    f.write(text_result)
    
```
