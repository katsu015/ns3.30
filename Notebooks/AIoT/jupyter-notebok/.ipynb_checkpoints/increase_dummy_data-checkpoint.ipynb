```python
df_original = pd.DataFrame(columns=['width', 'height', 'area', 'depth_sum', 'depth_sum/area','length', 'depth_median', 'depth_mean', 'ratio','r_median',
     'b_median', 'g_median', 'r_mean', 'b_mean', 'g_mean', 'r_sum', 'b_sum', 'g_sum','category' ,'y'])
df = pd.read_csv('base.csv', index=False)

for _ in range(0, len(df)*100):
    select_num = random.randint(0, 8)
    data = df.loc[select_num]
    width_bias = random.randint(-3, 3)
    height_bias = random.randint(-3, 3)
    depth_sum_bias = random.randint(-30000, 30000)
    depth_median_bias = random.randint(-2, 2)
    depth_mean_bias = random.randint(-10, 10)
    color_median_bias = random.randint(-15, 15)
    color_mean_bias = random.randint(-15, 15)
    color_sum_bias = random.randint(-450000, 450000)
    


    width = data.width + width_bias
    height = data.height + height_bias
    area = width * height
    length = width+height
    depth_sum = data.depth_sum + depth_sum_bias
    depth_sum_area = depth_sum/area
    depth_median = data.depth_median + depth_median_bias
    depth_mean = data.depth_mean + depth_mean_bias
    w_h = [width, height]
    ratio = max(w_h)/min(w_h)
    r_median = data.r_median + color_median_bias
    b_median = data.b_median + color_median_bias
    g_median = data.g_median + color_median_bias
    r_mean = data.r_mean + color_mean_bias
    b_mean = data.b_mean + color_mean_bias
    g_mean = data.g_mean + color_mean_bias
    r_sum = data.r_sum + color_sum_bias
    b_sum = data.b_sum + color_sum_bias
    g_sum = data.g_sum + color_sum_bias
    y = data.y
    category = data.category

    df_original
    dataseries = {'width': width, 'height': height, 'area': area, 'depth_sum': depth_sum, 'depth_sum/area': depth_sum_area,
                  'length': length, 'depth_median': depth_median, 'depth_mean': depth_mean, 'ratio': ratio, 'r_median': r_median ,
                  'b_median': b_median, 'g_median': g_median, 'r_mean': r_mean, 'b_mean': b_mean, 'g_mean': g_mean, 'r_sum': r_sum, 
                  'b_sum': b_sum, 'g_sum', 'y': y}
    df_new = pd.DataFrame.from_dict(dataseries, orient='index').T
    df_original = pd.concat([df_original, df_new], sort=False)
    df_original.to_csv('data/csv/train.csv', index=False)
```

```python
df_original.drop(['width', 'height'], axis=1)
```

```python
for _ in range(0, 10000):
    select_num = random.randint(0, 24)
    data = df.loc[select_num]
    width = data.length/(1+data.ratio)
    height = data.length - width
    width_bias = random.randint(-10, 10)
    height_bias = random.randint(-10, 10)
    depth_sum_bias = random.randint(-50000, 50000)
    depth_median_bias = random.randint(-4, 4)
    depth_mean_bias = random.randint(-12, 12)


    width = width + width_bias
    height = height + height_bias
    length = width + height
    area = width*height
    depth_sum = data.depth_sum + depth_sum_bias
    depth_sum_area = depth_sum/area
    depth_median = data.depth_median + depth_median_bias
    depth_mean = data.depth_mean + depth_mean_bias
    y = data.y

    df_original
    dataseries = {'width': width, 'height': height, 'area': area, 'depth_sum': depth_sum, 'depth_sum/area': depth_sum_area,
                  'length': length, 'depth_median': depth_median, 'depth_mean': depth_mean, 'ratio': data.ratio, 'y': y}
    df_new = pd.DataFrame.from_dict(dataseries, orient='index').T
    df_original = pd.concat([df_original, df_new], sort=False)
```
