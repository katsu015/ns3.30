```python
import json
import numpy as np
import cv2 
from PIL import Image
import re
import pandas as pd
import glob
import random
```

```python
json_open = open('json/fearure_json.json', 'r')
json_load = json.load(json_open)
```

```python
df = pd.DataFrame.from_dict(json_load, orient='index')
```

```python
def make_ratio(data):
    return data['width']/data['height']

def col_median(data):
    return (data[0]+data[1])/2

def col_range(data):
    return data[1]-data[0]
```

```python
df['ratio'] = df.apply(make_ratio, axis=1)
df['depth_median'] = df['depth'].apply(col_median)
df['depth_range'] = df['depth'].apply(col_range)
```

```python
def add_y(data):
    return data.name
```

```python
df['y'] = df.apply(add_y, axis=1)
df['length'] = df['width'] + df['height']
```
