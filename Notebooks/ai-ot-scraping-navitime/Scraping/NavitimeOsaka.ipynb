```python
from bs4 import BeautifulSoup
import requests
import json


url = 'https://www.navitime.co.jp/maps/poi?code=02047-638427'
response = requests.get(url)         #requestsを使って、webから取得

soup = BeautifulSoup(response.content) #要素を抽出

# soup = soup.find('div', class_='spot')
# print(soup)
# #a要素を全て取得
a_list = soup.find_all("a")

for a in a_list:
    #取得したaタグを表示
    #print(a.string)
    if 'href' in a.attrs:
     #aタグのhref属性値を抽出する
     href = a.attrs['href']
     #aタグの文字列と、そこからリンクしているURLを表示
     print(a.string, '-->', href)


# beautiful soupではscriptが取得できない

```

```python
from selenium import webdriver
import urllib.parse
import urllib.request, urllib.error
from selenium.common.exceptions import NoSuchElementException
```

```python
Path = r"C:\Users\shuto\chromedriver_win32\chromedriver"
driver = webdriver.Chrome(Path)

# driver pathを通す方法 chromedriver.exeのプロパティのパスをコピー　＋ \chromedrive



```

```python
driver.get('https://www.navitime.co.jp/maps/aroundResult?lat=34.827541&lon=135.700427&category=0516&type=spot&radius=1000')
driver.implicitly_wait(15)


coutText = driver.find_element_by_xpath("//*[contains(text(), '件の候補があります')]")
companyCount = coutText.text[0:3]   #検索ヒット数をcompanyCountで定義」
offSet = 20

print(companyCount)  

#この書き方だと１件のdataItemが取得できる　ループで回す必要あり
# dataItem = driver.find_element_by_xpath("//*[@id='around-content']/div/section/div/ul/li[1]/div").get_attribute("data-item")
# print(dataItem)  

# #ループ版のコード  
dataItems = []
for i in range(offSet+1):
    i = i+1
    index = str(i)
    
    try:
        url = driver.find_element_by_xpath("//*[@id='around-content']/div/section/div/ul/li["+ index +"]/div").get_attribute("data-item")
        item = urllib.parse.unquote(url)
        dataItems.append(item)
    except NoSuchElementException:
        print('例外が検出されました')
        continue
    
    print(dataItems)

# item = urllib.parse.unquote(driver.find_element_by_xpath("//*[@id='around-content']/div/section/div/ul/li["+ index +"]/div").get_attribute("data-item"))
    


```

```python
def is_json(myjson):
    try:
        json_object = json.loads(myjson)
    except ValueError as e:
        return False
    return True


def tryError(dictionary, key):
    try:
        keyValue = dictionary[key]
    except KeyError:
        keyValue = "掲載なし"
    return keyValue
        

    
nameList = []
addressList = []
phoneList = []



#dataItemsは連想配列ではなくて json形式ではないか
for dataItem in dataItems:
#     print(is_json(dataItem))   #json形式だということが判明
    d = json.loads(dataItem)   #json形式を辞書型に変換
#     print(d)
#     print("\n")
    name = tryError(d, 'name')
    nameList.append(name)
    address = tryError(d, 'address_name')
    addressList.append(address)
    phone = tryError(d, 'phone')
    phoneList.append(phone)
    
#     print(type(d))      #型の確認
#     name = d['name']
#     address = d['address_name']
#     phone = d['phone']
#     print(name,address,phone)
    


```
