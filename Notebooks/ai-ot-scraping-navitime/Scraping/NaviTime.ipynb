```python
import re
driver.get('https://www.navitime.co.jp/maps/aroundResult?lat=34.68696888888889&lon=135.60290111111112&category=0516&type=spot&radius=3000')
driver.implicitly_wait(45)

offSet = 20 #一ページ当たりの表示件数
```

driver.get(～)に検索した直後のリンクをコピペする

```python
countText = driver.find_element_by_xpath("//*[contains(text(), '件の候補があります')]")
print(countText.text)
companyCount = re.sub("\\D", "", countText.text)
companyCount = int(companyCount)
print(companyCount)
```

ヒットした件数をcampanyCountに格納

```python
roopCount = companyCount/offSet
roopCount = math.ceil(roopCount)
roopCount
```

<br>roopCount回ループさせる
<br>リンクが以下の通りになっているので
<br>https://www.navitime.co.jp/maps/aroundResult?lat=34.82754111111111&lon=135.70042694444444&category=0516&type=spot&radius=1000
<br>１回目は20件取得し　その後ループで回す
<br> 2回目以降のループ
<br> https://www.navitime.co.jp/maps/aroundResult?lat=34.82754111111111&lon=135.70042694444444&category=0516&type=spot&offset=20&radius=1000
<br>offsetの値が20ずつ増加する

```python
def is_json(myjson):
    try:
        json_object = json.loads(myjson)
    except ValueError as e:
        return False
    return True

def tryError(dictionary, key):
    try:
        keyValue = dictionary[key]
    except KeyError:
        keyValue = "掲載なし"
    return keyValue
           
nameList = []
addressList = []
phoneList = []
dataItems = []
```

<br>is_jsonはjson形式なのかを判別する
<br>tryErrorはkeyに該当するvalueがない場合掲載なしという文字列を返す
<br>nameList,addressList,phoneListにそれぞれ会社名、住所、電話番号を格納していく
<br>dataItemsは取得したjsonコードをすべてを格納する配列

# 1ページ目の取得 1ページ目取得完了ログが出るまで待機

```python
for i in range(offSet+1):
    i = i+1
    index = str(i)
    
    try:
        url = driver.find_element_by_xpath("//*[@id='around-content']/div/section/div/ul/li["+ index +"]/div").get_attribute("data-item")
        item = urllib.parse.unquote(url)
        dataItems.append(item)
        #print(item)
    except NoSuchElementException:
        print('例外が検出されました')
        continue
print("1ページ目取得完了")
```

```python
for dataItem in dataItems:
    d = json.loads(dataItem)   #json形式を辞書型に変換
    name = tryError(d, 'name')
    nameList.append(name)
    address = tryError(d, 'address_name')
    addressList.append(address)
    phone = tryError(d, 'phone')
    phoneList.append(phone) #それぞれ格納

print(nameList)# nameList check
```

# 2列目以降はループで記述していく

```python
def getOffSetData():
    for i in range(offSet+1):
        i = i+1
        index = str(i)
        print(index)
        try:
            url = driver.find_element_by_xpath("//*[@id='around-content']/div/section/div/ul/li["+ index +"]/div").get_attribute("data-item")
            item = urllib.parse.unquote(url)
            dataItems.append(item)
            print('item append')
        except NoSuchElementException:
            print('例外が検出されました')
            continue
```

取得完了ログが出るまで待機

```python


for i in range(roopCount):
  if i == 0:
    continue
  else:
    num = i*offSet
    num = str(num)
#     driver.get("https://www.navitime.co.jp/maps/aroundResult?lat=34.82754111111111&lon=135.70042694444444&category=0516&type=spot&offset="+ num +"&radius=3000")
    driver.get("https://www.navitime.co.jp/maps/aroundResult?lat=34.68696888888889&lon=135.60290111111112&category=0516&type=spot&offset="+ num +"&radius=2000")
    driver.implicitly_wait(45)
    getOffSetData()           
    time.sleep(random.randint(1,100))
    

print("取得完了")
```
